### reading file

```python
def read_csv(filename):
    x1 = []
    x2 = []
    x3 = []
    x4 = []
    y = []
    with open(filename) as f:
        reader = csv.reader(f)
        next(reader)
        for row in reader:
            x1.append(int(row[0]))
            x2.append(int(row[1]))
            x3.append(int(row[2]))
            x4.append(int(row[3]))
            y.append(int(row[4]))
        return x1,x2,x3,x4,y
if __name__ == '__main__':
    x1,x2,x3,x4,y = read_csv('LR.csv')
```

```python
x = np.array(np.vstack((x1,x2,x3,x4)))
print(x)
```

```python
def scatter_plot(x1,x2,x3,y):
    plt.style.use("ggplot")
    plt.rcParams["figure.figsize"] = (12,10)
    plt.scatter(x1,x2,(x3),y)
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.show()
```
